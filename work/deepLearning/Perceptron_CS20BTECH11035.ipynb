**PERCEPTRON CLASS**

```python
#created a class Perceptron which has attributes number of iterations(k),learning rate(lr), weights(weight) and bias(bias)

class Perceptron:
  def __init__(self,no_of_iters,learning_rate):
    self.k=no_of_iters
    self.lr=learning_rate
    self.weight=None
    self.bias=None


  #perceptron learning algorithm
  #train_perceptron returns number of iterations required.
  def train_perceptron(self,X,Y):
    self.weight= torch.zeros(X.shape[1])
    self.bias=0
    for j in range(self.k):
      no_of_changes=0
      for i in range(X.shape[0]):
          yp=0.0
          if X[i].dot(self.weight) + self.bias>0:
            yp=1.0
          if yp!=y[i]:
            #update weight and bias
            self.weight= self.weight+ self.lr*(y[i]-yp)*X[i]
            self.bias= self.bias+ self.lr*(y[i]-yp)
            no_of_changes=no_of_changes+1
      if no_of_changes==0:
        break
    return (j+1)
    
```

**DATASET CREATION**

```python
#dataset creation
D1=torch.normal(0.0, 1.0,size=[500,2])
D2=torch.normal(5.0, 1.0,size=[500,2])
X=torch.empty(1000,D1.size(1)).fill_(1.0)
X[::2,:2] = D1
X[1::2,:2] = D2


y=torch.empty(1000,)
y[::2] = 1.0
y[1::2] = 0.0
```

**PERCEPTRON TRAINING**
